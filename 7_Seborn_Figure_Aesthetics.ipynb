# Seaborn: statistical data visualization
## Controlling figure aesthetics in seaborn

Hi guys,<br>
Welcome back to the seaborn: a statistical data visualization section. We are going to talk about one of the most important aspect in the data presentation and report writing which is figure aesthetics. This very important aspect if widely ignored by the early stage data scientist which could significantly reduce the impact of all your efforts at the end. Always pay attention to the figure aesthetics!<br>


In the previous lectures, we have seen a couple of times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## Seaborn figure styles
### `set_style()`
We can set particular styles using `set_style()`. `shift+tab` for doc string<br>
There are five preset seaborn `themes: darkgrid, whitegrid, dark, white, and ticks`. They are each suited to different applications and personal preferences. The default theme is `darkgrid`. <br>


```python
# default countplot 
sns.countplot(x='sex',data=tips)
```

```python
# setting white background
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
# Passing ticks will add the ticks on the plot
sns.set_style('ticks') # try darkgrid, whitegrid
sns.countplot(x='sex',data=tips)#,palette='deep')
```

## Removing axes spines
### `despine()`
We can remove spine using **`despine()`** method. <br>
By default:  
`top=True, right=True, left=False, bottom=False`

```python
sns.countplot(x='sex',data=tips)
sns.despine() # top and right is True by default
```

```python
# if we want to remove left and bottom, pass left = True, bottom = True
sns.countplot(x='sex',data=tips)
sns.despine(left=True, bottom=True)
```

### Figure size and Aspect
We can simply use matplotlib's **`plt.figure(figsize=(width,height)` ** to change the size of most seaborn plots.

We can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:<br>

&#9758; In the 2nd line of the code below, when seaborn calls matplotlib, it actually set the figure size to the given size for seaborn.<br>
So, we can actually use our matplotlib knowledge in combination with seaborn to set the figure size and aspect ratio.

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# Another example with grid type plot
sns.lmplot(x='total_bill',y='tip',data=tips,size=5,aspect=2)
```

## Scaling plot elements
### `set_context()`
The `set_context()` allows us to override default parameters:<br>
Parameter context : `dict, None, or one of {paper, notebook, talk, poster}`

```python
sns.set_context('paper',font_scale=3)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
