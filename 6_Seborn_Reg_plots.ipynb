```python
tips.head()
```

### `lmplot()`

`lmplot()` allows us to display linear models with seaborn. The method, plot data and regression model fits across a FacetGrid. This function actually combines `regplot()` and `FacetGrid`.<br>

<font style="font-size:10px;color:green;">
Understanding the difference between `regplot()` and `lmplot()` can be a bit tricky. In fact, they are closely related, as `lmplot()` uses `regplot()` internally and takes most of its parameters. `lmplot()` is a figure-level function and creates its own figure, which is managed through a FacetGrid.</font> Explore more on [`lmplot()`](http://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot)

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```

In the plot above, we basically have a scatter plot with linear fit on top of it. <br>

Let's pass `'sex'` as `'hue'` argument.

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,
           hue = 'sex')
#sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

In the plot above, `hue = sex` conveniently split up the plots, as well as color the `hue` based on the categorical features. <br>
Now, we have two scatter plots and tow linear fits with different color on top of each other. 
We observe that, as far as total_bill vs tip are concerned, the male and female have similar linear fit on the data!

We can pass the [matplotlib style parameters](https://matplotlib.org/api/markers_api.html). For example, let's pass the list of markers for male and female. <br>
<font style="font-size:10px;color:green;">Its good to know and don't worry about remembering these parameters. You may not use them often and if needed, you can always check the reference notebook or consult the official website of matplotlib and seborn.</font>

```python
# passing the list of markers
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,
           hue = 'sex', 
           markers = ['o','v'])
```

```python
# Changing the marker size -- passing the dict here to scatter_kws
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,
           hue = 'sex',
           markers = ['o','v'],
           scatter_kws = {'s':100})
```

```python
# pelette = coolwarm
sns.lmplot(x = 'total_bill',y = 'tip', data = tips,
           hue = 'sex',
           markers = ['o','v'],
           scatter_kws = {'s':100},
           palette = 'coolwarm')
```

### Using Grid
We can add more variable separation through **columns and rows with the use of a grid**. <br>
Just indicate using the col or row arguments in `lmplot()`:

```python
# let's set sex as col.
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,
           col ='sex')
```

```python
# let's set sex as col. and time as row!
sns.lmplot(x = "total_bill",y = "tip", data = tips,
           row = "sex",
           col = "time")
```

```python
# col, row and hue together!
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,
           col = 'day',
           hue = 'sex',
           palette = 'coolwarm')
```

### Aspect ratio and Size

Seaborn figures can have their size and aspect ratio adjusted with the **size** and **aspect** parameters:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
           col='day',
           hue='sex',
           palette='coolwarm',
           aspect=0.6,
           size=8)
```
